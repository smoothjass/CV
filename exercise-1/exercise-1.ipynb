```python
import cv2
import matplotlib.pyplot as plt
```

load an image, change the color channel because cv uses BGR per default, and show it

```python
img = cv2.imread("7zdgks.jpg", 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

print the image array, its shape and its dimension to be able to see what is happening later during the processing

```python
print(img)
print(img.shape)
print(img.ndim)
```

rotate the img, convert it to grayscale and show it again, this time with the colormap set to gray

```python
img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img, cmap="gray")
```

print the info again to compare

```python
print(img)
print(img.shape)
print(img.ndim)
```

the values in the array were rearranged to rotate the picture. additionally, the third dimension, which defines color, was removed
